```python
from sklearn.tree import DecisionTreeClassifier 
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=2)
classifier.fit(x_train,y_train)
```

```python
#for the graph tre
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(classifier,filled=True)
```

```python
#predicit the data 
ypred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(ypred,y_test)
print(classification_report(ypred,y_test))
```

NOW I  HAVE TO DO PRE PREUNING USING HYPERPARAMETER TUING

```python
parameter={
    'criterion':['gini','entropy','log_looss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6,7,10],
    'max_features':['sqrt','log2','auto']
}
```

```python
from sklearn.model_selection import GridSearchCV
clf= DecisionTreeClassifier()
model=GridSearchCV(clf,param_grid=parameter,cv=5,scoring='accuracy',verbose=3)
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
ypred=model.best_estimator_.predict(x_test)
ypred
```

```python
accuracy_score(ypred,y_test)
```
